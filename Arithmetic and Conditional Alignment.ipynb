# Unique Value and ValueCounts

```python
import pandas as pd
car_data = {'Model Name':['valiant',
                         'duster 360',
                         'merc 2400',
                         'merc 230',
                         'merc 280',
                         'merc 280c',
                         'merc 450SE',
                         'merc 450SL',
                         'merc 450SLC',
                         'cadillac fleetwood',
                         'lincoln continental',
                         'chrysler imperial',
                         'fiat 128',
                         'honda civic',
                         'toyota corolla'],
           'Gear':[3,3,4,4,5,4,3,3,3,3,3,3,4,4,4],
           
           'cylinder':[6,8,4,4,6,6,8,8,8,8,8,8,4,4,4]}
car_df=pd.DataFrame(car_data)
car_df
```

```python
print(f"Unique values present in gear column are: {car_df['Gear'].unique()}")
```

```python
unique_gear=pd.unique(car_df.Gear)
unique_cyl=pd.unique(car_df.cylinder)
print(f"Unique values present in Gear column are : {unique_gear}")
print(f"Unique values present in Cylinder column are : {unique_cyl}")
```

```python
gear_count=pd.value_counts(car_df.Gear)
cyl_count=pd.value_counts(car_df.cylinder)
g_count=car_df['Gear'].value_counts()
c_counts=car_df['cylinder'].value_counts()
print('--------Output from first method--------')
print(gear_count)
print(cyl_count)
```

```python
name_count=pd.unique(car_df['Model Name'])
gear_count=pd.unique(car_df.Gear)
cyl_count=pd.unique(car_df.cylinder)
name_unique=0
gear_unique=0
cyl_unique=0

for item in name_count:
    name_unique+=1
for item in gear_count:
    gear_unique+=1
for item in cyl_count:
    cyl_unique+=1
    
print(f"Number of unique values present in Model is: {name_unique}")
print(f"Number of unique values present in Gear is: {gear_unique}")
print(f"Number of unique values present in cylinder is: {cyl_unique}")
```

```python
print(len(name_count))
print(len(gear_count))
```

# Pivot Table 

```python
import pandas as pd 
import numpy as np
df = pd.DataFrame({'Name':['John','Bobby','Mina','Peter','Nicky'],
                  'Degree':['Masters','Graduate','Graduate','Masters','Graduate'],
                  'Age':[27,23,21,23,24]})
df
```

```python
table = pd.pivot_table(df,index = ['Name','Degree'])
#name and degree are assigned as index.
table
```

```python
table = pd.pivot_table(df,values='Name',index=['Degree','Age'],
                      columns = ['Degree'],aggfunc = np.sum)
table
```

# Pandas Plotting

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
ts = pd.Series(np.random.randn(1000),index = pd.date_range('1/1/2000',periods = 1000))
ts = ts.cumsum()
ts.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
ts = pd.Series(np.random.randn(1000),index = pd.date_range('1/1/2000',periods = 1000))
df = pd.DataFrame(np.random.randn(1000,4),
index = ts.index,columns = list('ABCD'))
df = df.cumsum()
plt.figure()
df.plot()
plt.show()
```

```python
df3 = pd.DataFrame(np.random.randn(1000,2),columns = ['B','C']).cumsum()
df3['A'] = pd.Series(list(range(len(df))))
df3.plot(x='A',y='B')
plt.show()
```

```python
#Bar Graph Plot

df3.iloc[5].plot.bar()
plt.axhline(0,color = 'k')
plt.show()
```

```python
#Histogram Plot

df4 = pd.DataFrame({'a':np.random.randn(1000) + 1,
                   'b':np.random.rand(1000),
                   'c':np.random.randn(1000) - 1},
                  columns=['a','b','c'])
plt.figure()
df4.plot.hist(alpha=0.5)
plt.show()
```

```python
#Box Plot

df = pd.DataFrame(np.random.rand(100,5),
                 columns=['A','B','C','D','E'])
df.plot.box()
plt.show()
```

```python
#Kernel Density Estimation Plot

ser = pd.Series(np.random.randn(1000))
ser.plot.kde()
plt.show()
```

```python
#Area Plot

df.plot.area()
plt.show()
```

```python
#Scatter Plot

df.plot.scatter(x='A',y='B')
plt.show()
```

```python
#Pie Plot

series = pd.Series(3*np.random.rand(4),
                  index = ['a','b','c','d'],name='series')
series.plot.pie(figsize = (4,4))
plt.show()
```
