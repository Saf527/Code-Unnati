```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('My First Graph')
plt.show()
```

```python
ypoints = np.array([4, 7, 1, 10])
plt.plot(ypoints, marker = '*',color='g',linestyle='dashed')
plt.show()
```

```python
x=np.arange(0,10,0.2)
y=np.sin(x)
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=np.cos(x)
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.2)
y=np.tan(x)
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.show()
```
